## **Punto 1 - Viga compuesta 1**
----

Con la figura mostrada determine las componentes de la reacción de los soportes. Adjunte una foto en el cuaderno de jupyter o en BN del sistema de ecuaciones al que llegó.


<img src="./img/Ejercicio_1.png"  height='300'/>

![Alt text](image-11.png)

```python
# Importar librerías 
import sympy
B_y = sympy.Symbol('B_y')
B_x = sympy.Symbol('B_x')
C_y = sympy.Symbol('C_y')

ec1_a = sympy.Eq(0,-8+C_y*2)
ec2_a = sympy.Eq(0,B_y-8+C_y)
ec3_a = sympy.Eq(0,B_x)

sol1a = sympy.solve([ec1_a,ec2_a, ec3_a],[C_y,B_y, B_x])
sol1a
```

![Alt text](image-12.png)

```python
#Definir variables
A_x = sympy.Symbol('A_x')
A_y = sympy.Symbol('A_y')
M_a = sympy.Symbol('M_a')

#Valor conocido
B_y_val = sol1a[B_y]
B_x_val = sol1a[B_x]

ec2a = sympy.Eq(0,-10*(4/5)+A_y-B_y_val)
ec2b = sympy.Eq(0,-10*(3/5)+B_x_val + A_x)
ec2c = sympy.Eq(0,M_a+10*(4/5)*2-A_y*4)

solb = sympy.solve([ec2a,ec2b,ec2c],[A_y,A_x,M_a])
solb

print("Reacción vertical en A (A_y) en N:", solb[A_y])
print("Reacción horizontal en A (A_x) en N:", solb[A_x])
print("Momento en A (M_a) en N*m:", solb[M_a])
print("Reacción horizontal en B (B_x) en N:", sol1a[B_x])
print("Reacción vertical en B (B_y) en N:", sol1a[B_y])
```

## **Punto 2 - Ejercicio viga compuesta**
----

La figura muestra un bastidor que soporta 50 kg. Determine las componentes horizontal y verticual de la rección en A y la fuerza en C. Adjunte una foto en el cuaderno de jupyter o en BN del sistema de ecuaciones al que llegó.


<img src="./img/Ejercicio_2.png"  height='300'/>

![Alt text](image-8.png)

```python
import sympy
#Definimos variables simbólicas
D_x = sympy.Symbol('D_x')
D_y = sympy.Symbol('D_y')
T = sympy.Symbol('T')
F_p = sympy.Symbol('F_p')

masa = 50
g = 9.81
F_p = T
F_p = masa*g


ec1 = sympy.Eq(D_x,F_p)
ec2 = sympy.Eq(D_y,F_p)

sol = sympy.solve([ec1,ec2],[D_x,D_y])
sol
```

![Alt text](image-9.png)

```python
#Hallar la reacción vertical y horizontal en A

#Definimos variables simbólicas
F_bc = sympy.Symbol('F_bc')
A_x = sympy.Symbol('A_x')
A_y = sympy.Symbol('A_y')

# Valores conocidos
D_x_val = sol[D_x]  
D_y_val = sol[D_y] 

#Sistema 3 x 3
ec1_ = sympy.Eq(0,F_bc*0.6+D_x_val*0.9-D_y_val*1.2)
ec2_ = sympy.Eq(0,A_y-D_y_val)
ec3_ = sympy.Eq(0,A_x-D_x_val-F_bc)

sol1 = sympy.solve([ec1_,ec2_,ec3_],[F_bc,A_y,A_x])
sol1
```

![Alt text](image-10.png)

```python
#Hallar reacciones en C

#Definimos variables simbólicas
C_y = sympy.Symbol('C_y')
C_x = sympy.Symbol('C_x')

# Valor conocido
F_bc_val = sol1[F_bc]  
 
#Sistema de ecuaciones
ec1_f = sympy.Eq(0,C_y)
ec2_f = sympy.Eq(0,F_bc_val+C_x)

sol2 = sympy.solve([ec1_f,ec2_f],[C_y,C_x])
sol2
```

```python
print("Reacción vertical en A (A_y) en N:", sol1[A_y])
print("Reacción horizontal en A (A_x) en N:", sol1[A_x])
print("Reacción vertical en C (C_y) en N:", sol2[C_y])
print("Reacción horizontal en C (C_x) en N:", sol2[C_x])
```
